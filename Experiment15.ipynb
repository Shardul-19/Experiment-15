Part-A Data Normalisation

```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'price':[55000,20000,30000,2000,8000,45000],
    'Unit_Sold':[25,60,35,80,50,15],
    'Discount':[10,15,12,20,18,8]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1. Min-Max Normalisation
# Min-Max Normalisation for Price

df['Price_MinMax'] = (df['price'] - df['price'].min()) / (df['price'].max() - df['price'].min())
print(df[['Product','price','Price_MinMax']])
```

```python
#Normalisation Multiple Columns

cols = ['price','Unit_Sold','Discount']
df_norm = df[cols] - df[cols].min()
df_norm = df_norm / (df[cols].max() - df[cols].min())
df_norm
```

```python
#2.Z-score Normalisation
df['Units_Zscore'] = (df['price'] - df['price'].mean()) / df['price'].std()
print(df[['Product','price','Units_Zscore']])
```

```python
#3.Decimal Scaling
df['Price_Decimal'] = df['price'] / 100000
```

```python
import pandas as pd
import numpy as np
data = {
'Order_ID':[101,102,103,104,105,106],
'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
'Order_Value':[4500,2300,3200,5200,1800,4100]
}

df1 = pd.DataFrame(data)

print(df1)
```

```python
#METHOD1: Label Encoder

from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

# Fit and transform the 'Customer_Gender' column
df1['Gender_Encoded'] = le.fit_transform(df1['Customer_Gender'])

# Display the DataFrame with the new encoded column
print(df1[['Customer_Gender', 'Gender_Encoded']])
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le_city = LabelEncoder()

# Fit and transform the 'City' column
df1['City_Encoded'] = le_city.fit_transform(df1['City'])

# Display the DataFrame with the new encoded column
print(df1[['City', 'City_Encoded']])
```

```python
#method 2: one-hot encoding
df_encoded1=pd.get_dummies(df1,columns=['Customer_Gender'])
print(df_encoded1)
```

```python
#method 2: one-hot encoding
df_encoded1=pd.get_dummies(df1,columns=['Customer_Gender', 'City'], dtype=int)
print(df_encoded1)
```
